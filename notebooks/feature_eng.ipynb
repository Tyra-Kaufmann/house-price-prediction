```python
from pgeocode import Nominatim
```

```python
df = pd.read_csv('../data/housing.csv')

df.head()
```

```python
zips = df['zipcode'].to_list()
zips = [str(zipcode) for zipcode in zips]

geo = Nominatim('us')
results = geo.query_postal_code(zips)

df['city'] = results['place_name'].to_list()
```

```python
df['house_age'] = 2025 - df['yr_built']
```

```python
df['was_renovated'] = df['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df['renovation_age'] = df.apply(
    lambda row: 2025 - row['yr_renovated'] if row['yr_renovated'] > 0 else row['house_age'],
    axis=1
)
```

```python
df['pct_basement'] = df['sqft_basement'] / df['sqft_living']
```
